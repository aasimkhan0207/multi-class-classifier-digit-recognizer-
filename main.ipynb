```python
from keras.models import Sequential
```

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
np.random.seed(1)
from keras.layers import Dense
```

```python
data  = pd.read_csv('data/train.csv')
```

```python
data.head(3)
```

```python
#### image has been flattened into a 784-column vector. So, to display the images, we need to reshape into a 28x28 matrix.
#### NOTE :not changing actual data this is for sake of visualization
```

```python
X = data.loc[:,'pixel0':] #[row, col]
y = data.loc[:,'label']
```

```python
X.columns
```

```python
X.shape
```

```python
# one horizontal line in csv file represent one image 
for n in range(1,10): # 9 times
    plt.imshow(X.iloc[n].values.reshape((28,28)),cmap='gray') # converting to matrix 28x28  and display 
    plt.title(y.iloc[n])
    plt.show()
```

#### We can there there are ten classes

```python
# Create training data set
X_train = X[:40000]
y_train = y[:40000]
y_train = pd.get_dummies(y_train)
```

```python
y_train # 0-9 ten possible classes
```

```python
## for cross validation
X_test = X[40000:42000]
y_test = y[40000:42000]
y_test = pd.get_dummies(y_test)
```

```python
print('no. of training data sets: {}'.format(X_train.shape[0]))
print('no. of testing data sets: {}'.format(X_test.shape[0]))
```

### Defining and Training model

<ul>
<li>input layer has to correspond to the number of features in the input vector</li>
<li>number of layers in the final layer has to correspond to the number of categories. </li>
<li>we need 784 inputs (28x28) and an output for each of the 10 digits. </li>
</ul>

```python
model = Sequential()
model.add(Dense(32, activation='relu', input_dim=784))
model.add(Dense(16, activation='relu'))
model.add(Dense(10, activation='softmax')) # 10 for ten classes
```

```python
## For the loss function, 'binary_crossentropy' is a good choice for binary classification, 
## and 'categorical_crossentropy' is a good choice for multi-class classification.
```

```python
model.compile(optimizer='Adam',loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
## We need to decide on the number epochs. or iterations, 
## and the batch size, which represents the number of training examples that will be processed in each epoch.
```

```python
model.fit(X_train.values,
          y_train.values,
          epochs =20,
          batch_size = 64,
          verbose = 2,
          validation_data=(X_test.values, y_test.values)
         
         )
```
